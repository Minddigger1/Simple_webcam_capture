```python
import cv2

# Function to handle button clicks
def button_click(event, x, y, flags, param):
    global capture_requested, quit_requested

    # Check if the left mouse button was clicked
    if event == cv2.EVENT_LBUTTONDOWN:
        # Check if the click is within the "Capture" button region
        if 10 <= x <= 110 and 10 <= y <= 60:
            capture_requested = True
        # Check if the click is within the "Quit" button region
        elif 130 <= x <= 230 and 10 <= y <= 60:
            quit_requested = True

# Initialize the webcam
cap = cv2.VideoCapture(0)

# Initialize flags
capture_requested = False
quit_requested = False

# Create a window
cv2.namedWindow('Webcam')

# Set up the callback function for button clicks
cv2.setMouseCallback('Webcam', button_click)

while True:
    # Capture a frame from the webcam
    ret, frame = cap.read()

    # Draw the "Capture" button
    cv2.rectangle(frame, (10, 10), (110, 60), (0, 255, 0), thickness=cv2.FILLED)
    cv2.putText(frame, "Capture", (20, 45), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 0), 2)

    # Draw the "Quit" button
    cv2.rectangle(frame, (130, 10), (230, 60), (0, 0, 255), thickness=cv2.FILLED)
    cv2.putText(frame, "Quit", (160, 45), cv2.FONT_HERSHEY_SIMPLEX, 0.7, (255, 255, 255), 2)

    # Display the frame
    cv2.imshow('Webcam', frame)

    # Check if the "Capture" button was pressed
    if capture_requested:
        # Save the captured frame to a file
        captured_image = frame.copy()
        # Exclude the button areas from the captured image
        captured_image = captured_image[60:, :]
        cv2.imwrite('captured_image.jpg', captured_image)
        capture_requested = False

    # Check if the "Quit" button was pressed
    if quit_requested:
        # Release the webcam and close the window
        cap.release()
        cv2.destroyAllWindows()

        # Read the saved image from the file
        image = cv2.imread('captured_image.jpg')

        # Check if the image was successfully read
        if image is not None:
            # Display the captured image
            cv2.imshow('Captured Image', image)
            cv2.waitKey(0)
            cv2.destroyAllWindows()

        break

    # Check for key press to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
```
